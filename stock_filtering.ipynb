```python
import pandas as pd
import datetime as dt
import numpy as np
```

## Data Filtering

```python
all_stocks = pd.read_csv('data/all_stocks.csv')
```

```python
all_stocks
```

```python
all_stocks['Date'].min()
all_stocks['Date'].max()
```

```python
# Chaning the timeframe to be within the Twitter data timeframe
all_stocks = all_stocks[(all_stocks['Date'] >= '2016-04-01') & (all_stocks['Date'] <= '2016-09-28')]
```

```python
# Converting to a CSV file
all_stocks.to_csv("all_stocksV2.csv", header=True, index=False)
```

```python
# Reading in CSV file from above
all_filtered_stocks = pd.read_csv('all_stocksV2.csv')
```

```python
# Reading in the CSV lookup file
twitter_ticker_accounts = pd.read_csv('data/twitter_accounts_with_ticker_data.csv')
twitter_ticker_accounts
```

```python
# Performing a lookup for all stocks that we have twitter accounts for
ticker_symbols = twitter_ticker_accounts['Ticker Symbol'].str.lower()
tickers_v2 = all_filtered_stocks.stock.isin(ticker_symbols)
all_filtered_stocks['Match?'] = tickers_v2
all_filtered_stocks = all_filtered_stocks[all_filtered_stocks['Match?'] == True]
all_filtered_stocks
```

```python
# Creating a file stock CSV file
# Move to your data folder
all_filtered_stocks.to_csv('final_stock_file.csv')
```
