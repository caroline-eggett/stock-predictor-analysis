```python
import glob
import pandas as pd
```

```python
df_list = []
for file in files:
    try:
        stock = file.split('/')[-1].split('.')[0]
        df_list.append(pd.read_csv(file, header=None, sep=',').assign(stock=stock))
    except pd.errors.EmptyDataError:
        bad_df = pd.DataFrame()
    
print(df_list)
```
